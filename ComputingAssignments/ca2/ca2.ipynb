## Linear Transformations in 2D

Each column in `points` represents a point in $\mathbb{R}^2$. Create a scatter plot to visualize the points. In the cell below, the parameter `s = 2` specifies the size of the points in the plot and `c = [[0.75,0,0,1]]` specifies the color as an RGBA vector `[red,green,blue,opacity]`.

Now let's apply some geometrical transformations to the points in the data set, and plot the result data.

### Rotation Matrix

A rotation matrix is given by

$$
R_{\theta} = \begin{bmatrix} \cos(\theta) & -\sin(\theta) \\ \sin(\theta) & \cos(\theta) \end{bmatrix}
$$

The linear transformation corresponding to $R_{\theta}$ is rotation counterclockwise by $\theta$ radians.

```python
# construct the rotation matrix:
theta = np.pi/6
R = np.array([[np.cos(theta),-np.sin(theta)],
              [np.sin(theta),np.cos(theta)]])

# apply the matrix to the points with matrix multiplication:
points1 = R@points

# plot the collection of points (you don't need to worry about the syntax below,
# this will already be filled out in the susequent questions. You'll just focus
# on constructing the correct transformation matrix.
plt.scatter(points[0,:],points[1,:],s=2,c=[[0.75,0,0,0.1]]) # original points, dimmed
plt.scatter(points1[0,:],points1[1,:],s=2,c=[[0.75,0,0,1]]) # transformed points
plt.axis('equal'), plt.grid(True)
plt.show()
```

### Scaling Matrix

A scaling matrix is given by

$$
C_{a,b} = \begin{bmatrix} a & 0 \\ 0 & b \end{bmatrix}
$$

The linear transformation corresponding to $C_{a,b}$ is scaling by $a$ is the $x$-direction and scaling by $b$ in the $y$-direction.

```python
# construct the scaling matrix:
a = 3
b = 1/2
C = np.diag([a,b]) # this creates a 2x2 diagonal matrix with entries a,b along the diagonal

# apply the matrix to the points with matrix multiplication:
points2 = C@points

# plot the collection of points
plt.scatter(points[0,:],points[1,:],s=2,c=[[0.75,0,0,0.05]]) # original points, dimmed
plt.scatter(points2[0,:],points2[1,:],s=2,c=[[0.75,0,0,1]]) # transformed points
plt.axis('equal'), plt.grid(True)
plt.show()
```

### Shearing Matrix

A shear matrix is given by

$$
S_u = \begin{bmatrix} 1 & u \\ 0 & 1 \end{bmatrix}
$$

The linear transformation corresponding to $S_u$ is shearing by factor $u$ in the $x$-direction.

```python
# construct the scaling matrix:
u = 1
S = np.array([[1,u],
              [0,1]])

# apply the matrix to the points with matrix multiplication:
points3 = S@points

# plot the collection of points
plt.scatter(points[0,:],points[1,:],s=2,c=[[0.75,0,0,0.1]]) # original points, dimmed
plt.scatter(points3[0,:],points3[1,:],s=2,c=[[0.75,0,0,1]]) # transformed points
plt.axis('equal'), plt.grid(True)
plt.show()
```

### Reflection Matrix

A reflection matrix is given by

$$
F_x = \begin{bmatrix} 1 & 0 \\ 0 & -1 \end{bmatrix}
$$

The linear transformation corresponding to $F_x$ is a reflection across the $y$-axis.

Similarly, 

$$
F_y = \begin{bmatrix} -1 & 0 \\ 0 & 1 \end{bmatrix}
$$

is a reflection across the $y$-axis.

```python
# construct the reflection matrix:
F = np.array([[1,0],
              [0,-1]])

# apply the matrix to the points with matrix multiplication:
points4 = F@points

# plot the collection of points
plt.scatter(points[0,:],points[1,:],s=2,c=[[0.75,0,0,0.1]]) # original points, dimmed
plt.scatter(points4[0,:],points4[1,:],s=2,c=[[0.75,0,0,1]]) # transformed points
plt.axis('equal'), plt.grid(True)
plt.show()
```

### Combining Transformations with Matrix Multiplication

We can combine linear transformations using matrix multiplication. Recall the transformations in the product are applied from right-to-left, meaning that in the product below $R$ is applied first, then $C$, and finally $S$.

```python
# construct the scaling matrix:
points5 = R @ S @ C @ points

# plot the collection of points
plt.scatter(points[0,:],points[1,:],s=2,c=[[0.75,0,0,0.05]])
plt.scatter(points5[0,:],points5[1,:],s=2,c=[[0.75,0,0,1]])
plt.axis('equal'), plt.grid(True)
plt.show()
```
