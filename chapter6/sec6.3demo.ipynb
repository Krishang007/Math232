## Example 6.21

```python
P = np.array([[0.5, 0.4, 0.6],
              [0.2, 0.2, 0.3],
              [0.3, 0.4, 0.1]])
P
```

```python
x0 = np.array([0,1,0]) # intitial state vector (lion starts in reserve 2)
print(f"x0 = {x0}")

x1 = P@x0
print(f"x1 = {x1}")

x2 = P@x1
print(f"x2 = {x2}")
x3 = P@x2
print(f"x3 = {x3}")
x4 = P@x3
print(f"x4 = {x4}")
x5 = P@x4
print(f"x5 = {x5}")
x6 = P@x5
print(f"x6 = {x6}")
```

Can use a for loop to make it more robust:

```python
x = np.array([0,1,0]) # initial state vector (lion starts in reserve 2)
print(f"x0 = {x}")

for i in range(1,16):
    x = P@x
    print(f"x{i} = {x}")
```

The vector $[0.50420168 0.22689076 0.26890756] = [50\%,23\%,27\%]$ is the steady state solution.

## Example 6.24

The identity matrix can be created using the `np.eye` command.

```python
# We want to find the eigenvector corresponding to eigenvalue 1.
# This is the steady state solution.
rref(P-np.eye(3))
```

```python
v = np.array([1.875,0.84375,1]) # reading off the solution from the rref above
q = v/sum(v) # normalize it to a probability vector: sum of components must be 1
print(q) # display steady state solution
```

Note: this is the same vector we computed in Example 6.21 by using the for-loop and seeing what the $x_k$ converged to.
